```python
import pandas as pd
import numpy as np
```

```python
ecco_df = pd.read_csv('ECCO.csv', encoding='iso-8859-1')
ecco_df.head()
```

```python
ecco_df_extract = ecco_df[['RAMO', 'ANIO', 'INDICE DE SATISFACCION LABORAL']].copy()
ecco_df_extract.head()
```

```python
mean_satisfaction = ecco_df_extract.groupby(['RAMO', 'ANIO']).mean()
mean_satisfaction = mean_satisfaction.rename(columns={'RAMO':'ramo','ANIO':'anyo','INDICE DE SATISFACCION LABORAL':'promedio_ecco'})
mean_satisfaction.head()
```
