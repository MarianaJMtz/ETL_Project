```python
import pandas as pd
import numpy as np
```

```python
auditorias_df = pd.read_csv('Auditorias.csv', encoding='iso-8859-1')
auditorias_df.head()
```

```python
auditorias_df_ext = auditorias_df[['RAMO', 'EJERCICIO', 'AUDITORÍAS_REALIZADAS', 'OBSERVACIONES_DETERMINADAS']].copy()
auditorias_df_ext.head()
```

```python
auditorias_sanciones = auditorias_df_ext.groupby(['RAMO', 'EJERCICIO']).sum()
auditorias_sanciones = auditorias_sanciones.rename(columns={'AUDITORÍAS_REALIZADAS':'AUDITORIAS', 'OBSERVACIONES_DETERMINADAS':'SANCIONES'})
auditorias_sanciones['aud_san_ratio'] = auditorias_sanciones['AUDITORIAS'] / auditorias_sanciones['SANCIONES']
auditorias_sanciones = auditorias_sanciones.rename(columns={'RAMO':'ramo', 'EJERCICIO':'ejercicio','AUDITORIAS':'auditorias', 'SANCIONES':'sanciones'})
auditorias_sanciones
```

```python
auditorias_sanciones = auditorias_sanciones.reset_index()
auditorias_sanciones.dtypes
```
