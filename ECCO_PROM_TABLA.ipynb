```python
import pandas as pd
import numpy as np
```

```python
ecco_df = pd.read_csv('ECCO.csv', encoding='iso-8859-1')
ecco_df.head()
```

```python
ecco_df_extract = ecco_df[['RAMO', 'ANIO', 'INDICE DE SATISFACCION LABORAL']].copy()
ecco_df_extract.head()
```

```python
ecco_df_prom = ecco_df_extract.groupby(['RAMO', 'ANIO']).mean()
ecco_df_prom = ecco_df_extract.rename(columns={'INDICE DE SATISFACCION LABORAL':'PROMEDIO ECCO'})
ecco_df_prom.head()
```
